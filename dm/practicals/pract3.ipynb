## Imputing outliers

```python
def impute_outliers_df(df, lower_pct=0.05, upper_pct=0.95):
    df_imputed = df.copy()
    for col in df.select_dtypes(include=[np.number]).columns:
        lower = df[col].quantile(lower_pct)
        upper = df[col].quantile(upper_pct)
        median = df[col].median()

        mask = (df[col] < lower) | (df[col] > upper)
        df_imputed.loc[mask, col] = median
    return df_imputed

sales_df = impute_outliers_df(sales_df)
```

## Reducing dimensions using PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10, random_state=42)
sales_reduced = pd.DataFrame(pca.fit_transform(sales_df))
```

```python
sales_reduced
```

## Importing libraries for clustering

```python
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage,fcluster
from sklearn.metrics import silhouette_score,davies_bouldin_score, calinski_harabasz_score
from sklearn.neighbors import NearestNeighbors
```

## Using elbow method to get k values

```python
Z = linkage(sales_reduced, method='ward')

wcss = []
K_range = range(1, 11)

for k in K_range:
    labels = fcluster(Z, k, criterion='maxclust')

    total_wcss = 0
    for cluster_id in np.unique(labels):
        cluster_points = sales_reduced[labels == cluster_id].values
        centroid = cluster_points.mean(axis=0)
        total_wcss += ((cluster_points - centroid) ** 2).sum()

    wcss.append(total_wcss)

plt.plot(K_range, wcss, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method (Agglomerative Clustering) (on reduced df)")
plt.show()
```

## Using silhouette method to get k values

```python
silhouette_scores = []

K_range = range(2, 11)

for k in K_range:
    kmeans = AgglomerativeClustering(n_clusters=k,linkage="ward")
    labels = kmeans.fit_predict(sales_reduced)

    score = silhouette_score(sales_reduced, labels)
    silhouette_scores.append(score)

plt.plot(K_range, silhouette_scores, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Method (on reduced df)")
plt.show()
```

## Createing dendograph of clusters with found k values

```python
k = 4
threshold = (Z[-(k-1), 2] + Z[-k, 2]) / 2

plt.figure(figsize=(14, 7))
dendrogram(
    Z,
    truncate_mode='lastp',
    p=20,
    leaf_rotation=90,
    leaf_font_size=10,
    show_contracted=True
)

plt.axhline(y=threshold, color='red', linestyle='--', linewidth=1.5,
            label=f'Cut for k={k} clusters')

plt.title(f'Agglomerative Clustering Dendrogram (k={k})')
plt.xlabel('Sample Index / Cluster')
plt.ylabel('Ward Distance')
plt.legend()
plt.tight_layout()
plt.show()

labels = fcluster(Z, k, criterion='maxclust')
print(f"Clusters formed: {np.unique(labels)}")
print(f"Cluster sizes: {np.bincount(labels)[1:]}")
```

## Fine tunning `min_samples and `eps` values

```python
results = []
k_distances_s = {}
for min_samples in range(6, 40):
  result = {
      "eps":None,
      "silhouette score":None,
      "clusters":None,
      "noise pts":None,
      "noise percentage":None,
      "min_samples":None,
      "cluster 0 size":None,
      "cluster 1 size":None,
  }
  neighbors = NearestNeighbors(n_neighbors=min_samples)
  neighbors.fit(sales_reduced)
  distances, _ = neighbors.kneighbors(sales_reduced)

  k_distances = np.sort(distances[:, -1])[::-1]
  k_distances_s[min_samples] = k_distances
  for i in range(100):
    eps = k_distances[int(len(k_distances) * (i/100))]

    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(sales_reduced)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise    = list(labels).count(-1)
    if (n_clusters <= 1):
      continue

    if (n_noise/len(labels)*100 > 15):
      continue
    if n_clusters > 1 :
        score = silhouette_score(sales_reduced, labels)
        if score < 0:
          continue
        result["eps"] = eps
        result["silhouette score"] = score
        result["clusters"] = n_clusters
        result["noise pts"] = n_noise
        result["noise percentage"] = n_noise/len(labels)*100
        result["min_samples"] = min_samples
    unique, counts = np.unique(labels, return_counts=True)
    for u, c in zip(unique, counts):
        label_name = "Noise" if u == -1 else f"Cluster {u}"
        if (u == 0):
          result["cluster 0 size"] = c
        elif (u == 1):
          result["cluster 1 size"] = c
  if (result["eps"] != None):
    results.append(result)
```

```python
print(len(results))
values = pd.DataFrame(results).sort_values("silhouette score",ascending=False)
values
```

```python
print("Best silhouette score values")
best_row = values.iloc[0]
print(best_row)
```
